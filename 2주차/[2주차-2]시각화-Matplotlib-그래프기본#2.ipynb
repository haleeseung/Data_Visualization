### ○ CSV 파일 확인하기

```python
file = open('fine_dust.csv')
data = pd.read_csv(file, index_col='area')
data
```

### ○ 2018년 데이터 꺽은선 그래프

```python
plt.figure(figsize=(15, 4))
plt.plot(data['2018'], color='b', marker='o')
plt.xlabel('area')
plt.ylabel('micrometer')
plt.title('2018 Fine Dust Line Graph')
plt.grid()
plt.show()
```

### ○ EXAMPLE: 2016년~2019년 그래프 dsiplay
- 그래프 크기 =(15, 4)
- 서로 다른 marker, 선색깔로 구분
- 범례: 2016년~2019년
- Title: '미세먼지 변화량(2016년~2019년)' - x 축: 'Area', y 축: "Micrometer'

```python
plt.figure(figsize=(15,4))
plt.plot(data['2016'], color='r', marker='o', label='2016년')
plt.plot(data['2017'], color='b', marker='|', label='2017년')
plt.plot(data['2018'], color='g', marker='s', label='2018년')
plt.plot(data['2019'], color='k', marker='^', label='2019년')

plt.legend(loc=0) # 0 ~ 10

plt.xlabel('Area')
plt.ylabel('Micrometer')
plt.title('미세먼지 변화량(2016년~2019년)')
plt.grid()
plt.show()
```
