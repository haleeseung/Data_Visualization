```python
import matplotlib.pyplot as plt

# plt.grad() # 순서 확인
x = [1, 2, 3, 4]
y = [1, 4, 9, 16]
# plt.grid() # 순서 확인
plt.figure(figsize=(3, 2)) # 순서 확인. plt.plot보다 먼저 실행.
plt.grid()
plt.plot(x, y)
# plt.figure(figsize=(3, 2))
plt.show()
```

## ▣ 데이터 시각화 옵션

1. 제목(Title)
2. 범례(Legend)
3. 색상(Color)
4. 축(@label)
5. 선 모양(Linestyle, Marker)
6. 그래프 종류

### 1. 제목(Title)

```python
plt.figure(figsize=(4,3))
plt.title('Line Graph(라인 그래프)')
plt.plot(x, y)
```

### 2.범례(legend)

```python
plt.figure(figsize=(4,3))
plt.title('Line Graph(라인 그래프)')

# plt.legend() #순서 확인
plt.plot(x, label='x_data')
plt.plot(y, label='y_data')
plt.legend(loc=0) # 0~10

# plt.legend(bbox_to_anchor=(1,1)) # 좌표값으로 위치 다루기
```

○ Location code

Location string : Location code

‘best’ : 0

‘upper right’ : 1

‘upper left’ : 2

‘lower left’ : 3

‘lower right’ : 4

‘right’ : 5

‘center left’ : 6

‘center right’: 7

‘lower center’ : 8

‘upper center’: 9

‘center’ : 10

### 3. 색상(Color)

```python
plt.figure(figsize=(4,3))
plt.title('Line Graph(라인 그래프)')
plt.plot(x, label='x_data', color='r')
plt.plot(y, label='y_data', color='k')
plt.legend()
```

○ 색상 속성

속성값 : 색상

‘b’ : 파란색

‘g’ : 녹색

‘r’ : 빨간색

‘c’ : 청록색

‘m’ : 마젠타색

‘y’ : 노란색

‘k’ : 검은색

‘w’ : 흰색

### 4. 축(@label)

```python
plt.title('Line Graph')
plt.plot(x, label='x_data', color='r')
plt.plot(y, label='y_data', color='k')
plt.xlabel('Features', loc='right') # left, center, right
                                    # matplotlib 3.3 이후 적용 가능
# plt.xlabel('Features', color='b', fontsize=20) # 세부 기능들은 각자 학습
plt.ylabel('Predictive value', loc='top') #top, center, bottom
plt.legend()
plt.xlim(-0.5, 2.5) # x축 범위 설정
plt.ylim(-0.5, 9) #y축 범위 설정
# 그래프 내부 옵션과 외부 옵션들 사용 방법 주의
```

### 5. 선 모양(linestyle, marker)

```python
plt.title('Line Graph')
plt.xlabel('Fetaures', loc='right')
plt.ylabel('Predictive value', loc='top')
plt.plot(x, label='x_data', color='r', linestyle=':')
# 실선:'-'; 파선:'--'; 점쇄선:'-.', 점선:':'
plt.plot(y, label='y_data', color='k', marker='s', markersize
= '10')
# 'o', '+', ',', '.', '^', 's'
plt.legend()
```
